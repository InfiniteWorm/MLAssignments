# Theoritical solutions

## [Bayes Rule]

Let $A = $ "I have the disease", $A^C = $ "I don't have the disease, $B = $ "Test is positive", and $B^C = $ "Test is negative".

From the problem description, we know that $P(A) = \frac{1}{10000}$, and $P(B\mid A) = 0.99$ and $P(B^C \mid A^C) = 0.99$. We want to calculate $P(A\mid B)$, and that can be done by Bayes' Theorem : $$P(A\mid B) = \frac{P(B\mid A)P(A)}{P(B)}$$

We have all of the components to compute $P(A\mid B)$ except for $P(B)$, which can be calculated by :

$P(B) = (P(A) * P(B\mid A)) + (P(A^C) * (1-P(B^C\mid A^C)))$

$P(B) = \frac{1}{10000} * 0.99 + \frac{9999}{10000} * 1-0.99$ 

$P(B) = 0.010098$

Plugging in the variables, and we can calculate $P(A\mid B)$ : 

$$P(A\mid B) = \frac{P(B\mid A)P(A)}{P(B)} = \frac{0.99 * \frac{1}{10000}}{0.010098} \approx 0.0098 = 0.98\%$$

So, the probability of us actually having the disease is $0.98\%$

## [Correlation and Independence]

We can calculate $\mbox{cov}(X,Y)$ by :

$\mbox{cov}(X,Y) = E[XY] - E[X]E[Y]$

Since this is a uniform distribution, $E[X]$ is just the mean and we can calculate that by $\frac{1}{2}(a+b)$, with the $a$ and $b$ are the minimum and the maximum range of X, which is $-1$ and $+1$ respectively. Since $(a+b)$ is zero, then the mean is zero, and thus $E[X]$ is zero as well.

The PDF for X can be calculated by : $f_X(x) = \frac{1}{b-a} = \frac{1}{1-(-1)} = \frac{1}{2}$

Now if $Y = X^2$, then naturally $XY = X * X^2 = X^3$. Then we can calculate $E[XY]$ by : 

$E[XY] = \int_{-1}^{+1} X^3 * f_X(x) dx$

$E[XY] = \int_{-1}^{+1} X^3 * \frac{1}{2} dx = 0$

We can plug these values to the $\mbox{cov}(X,Y)$ : 

$\mbox{cov}(X,Y) = E[XY] - E[X]E[Y]$

$\mbox{cov}(X,Y) = 0 - 0E[Y]$

$\mbox{cov}(X,Y) = 0$

Q.E.D

## [Setting hyperparameters]

This is a beta distribution, so we can calculate both the mean and variance by : 

$$\mu = \frac{a}{a+b}$$ and $$\mbox{var}(\theta) = \frac{ab}{(a+b)^2(a+b+1)}$$

It is claimed in the description that $\mu = E[\theta] = m$ and $\mbox{var}[\theta] = v$

So for the mean function, we have : 

$$m = \frac{a}{a+b}$$

$$(a+b)m = a$$

$$(a+b) = \frac{a}{m}$$

$$b = a(\frac{1}{m}-1)$$

And for the variance function : 

$$v = \frac{ab}{(a+b)^2(a+b+1)}$$

$$v = \frac{a}{a+b}\frac{b}{a+b}\frac{1}{a+b+1}$$

$$v = \frac{m(1-m)}{a+b+1}$$

$$a+b = \frac{m(1-m)}{v}-1$$

$$\frac{a}{m} = \frac{m^2 - m^3 - v}{v}$$

$$a = \frac{m(m-m^2-v)}{v}$$

Since we have define $a$ in terms of $m$ and $v$, then we can define b as well by plugging a into this equation : 

$$b = a(\frac{1}{m}-1)$$

$$b = \frac{m-m^2-v}{v} - \frac{m(m-m^2-v)}{v}$$

$$b = \frac{(v+m^2-m)(m-1)}{v}$$

## [Plotting normal distributed points, 5 points]

Generate $1000$ points from 2D multivariate normal
distribution having mean $\mu = \left[
\begin{array}{c}
  1  \\
  1
\end{array}
\right]$ and covariance $\Sigma =
\left[
  \begin{array}{rr}
    0.1 & -0.05 \\
    -0.05&  0.2
  \end{array}
\right]
$.  Define the function $f({\bf x}, r) := \frac{({\bf x} - \mu)^{ \top } * \Sigma^{-1} *
  ({\bf x} - \mu) }{ 2} - r $. On a single plot, show the following:
* The level sets $f({\bf x}, r) = 0$ for $r=1, 2, 3$.
* Scatter plot of randomly generated points with points lying
outside $f({\bf x} , 3) = 0$ showing in black while points inside shown in
blue.
* Title of the plot showing how many points lie outside $f({\bf
    x}, 3) = 0$.
Submit your final plot as well as your implementation.

```python
import numpy as np
import matplotlib.pyplot as plt

# You can use, np.meshgrid() and np.contour to make your life easier

#The defined function
def fxr(x,mean,inverse,r):
    dif = np.subtract(x,mean);
    dift = np.transpose(dif);
    upper = dift * inverse * dif;
    result = (upper/2)-r;
    return result;

#Initialize values for N, mean, covariance matrix and its inverse
N = 1000;
mean = np.array([1,1]);
cov = np.matrix([[0.1,-0.05],[-0.05,0.2]]);
covinv = np.linalg.inv(cov);

#Generate N points from multivariate normal distribution
x = np.random.multivariate_normal(mean,cov,N)

#Initialize lists for blue and black points     
xlist = []
ylist = []
xBlist = []
yBlist = []

#Separate the the points based on value by testing the point [1,1] previously to be evaluated to a negative value 
# in f(x,3)=0, we concluded that all negative values derived from the points will be plotted blue, the rest black. 
for i in range(0,1000):
    c = np.array([x[i,0],x[i,1]])
    temp = fxr(c.reshape(2,1),mean.reshape(2,1),covinv,3)   
    if  temp < 0:
        xlist.append(x[i,0]);
        ylist.append(x[i,1]);
    else:
        xBlist.append(x[i,0])
        yBlist.append(x[i,1])
    
#set the resolution of the contour, and then initialize the lists for containing the function result value
reso = 200
Z = np.zeros(shape=(reso,reso))
Z2 = np.zeros(shape=(reso,reso))
Z3 = np.zeros(shape=(reso,reso))

#the values of x and y for the contour grid
xlist2 = np.linspace(-0.5, 2.5, reso)
ylist2 = np.linspace(-0.5, 2.5, reso)
X,Y = np.meshgrid(xlist2,ylist2);

#Building up our Z values
for i in range(0,reso):
    for j in range(0,reso):
        vector = np.array([X[i][j],Y[i][j]]).reshape(2,1)
        Z[i][j] = fxr(vector,mean.reshape(2,1),covinv,1)
        Z2[i][j] = fxr(vector,mean.reshape(2,1),covinv,2)
        Z3[i][j] = fxr(vector,mean.reshape(2,1),covinv,3)

#draw the contour plot and the scatterplot
plt.figure();
plt.title("Number of points outside f(x,3)=0 is %d" % (len(yBlist)))
plt.plot(xlist,ylist,".");
plt.plot(xBlist,yBlist,".", color ="black");
plt.contour(X,Y,Z,[0]);
plt.contour(X,Y,Z2,[0]);
plt.contour(X,Y,Z3,[0]);
plt.show();
```

## [Covariance and correlation, 5 points]
Load dataset0.txt ($X$) containing 1074 data points
each having 12 features related to US schools. Compute the covariance
and correlation matrix for $X$. Scale each feature
in $X$ between $[0, 1]$ to obtain a new dataset $Y$. Compute the
covariance and correlation matrices for $X$ and $Y$, and plot them (e.g. as colormaps).
What do you observe? Show a scatter plot of the pair of features in $Y$ having minimum
correlation, indicating in the title the feature indices and the
correlation value. Submit the plots, comments and your implementation.

```python
import numpy as np
import matplotlib.pyplot as plt
# You might want to load the data and analyze it first

#max-min normalization function, but we're normalizing by feature, not by schools
def normalize_data(dat):
    maxval = np.amax(dat,axis=0)
    minval = np.amin(dat,axis=0)
    result = np.zeros(shape=(dat.shape[0],dat.shape[1]))
    #result = (dat-minval.reshape(12,1))/(maxval-minval).reshape(12,1)
    cols = maxval.shape[0]
    datrows = dat.shape[0]
    for i in range(0,datrows):
        for j in range(0,cols):
            result[i][j] = (dat[i][j] - minval[j]) / (maxval[j]-minval[j])
    return result;

#load the data
X = np.loadtxt("dataset0.txt")

# covariance matrix for X
covX = np.cov(X)
# correlation matrix for X
corrX = np.corrcoef(X)

#scale X between [0,1], let this by Y
Y = normalize_data(X)

#compute covY and corrY
covY = np.cov(Y)
corrY = np.corrcoef(Y)

#plot these 4 matrices
f,((X1,X2),(Y1,Y2)) = plt.subplots(2,2)
plot1 = X1.imshow(covX)
X1.set_title("Covariance X")
plot2 = X2.imshow(corrX)
X2.set_title("Correlation X")
plot3 = Y1.imshow(covY)
Y1.set_title("Covariance Y")
plot4 = Y2.imshow(corrY)
Y2.set_title("Correlation Y")
plt.colorbar(plot1,ax=X1)
plt.colorbar(plot2,ax=X2)
plt.colorbar(plot3,ax=Y1)
plt.colorbar(plot4,ax=Y2)
plt.tight_layout()

# Since we want the correlation between features and not schools,
# we have to transpose the Y first, and then use the corrcoef function
featureY = Y.T
corrFeatureY = np.corrcoef(Y.T)

#find the minimum correlation in the correlation Y matrix
mincorrY = corrFeatureY.min()
#get the indices with the minimum correlation
ind = np.unravel_index(np.argmin(corrFeatureY,axis=None),corrFeatureY.shape)

#scatterplot the pair of features in Y that gives the smallest correlation
plt.figure()
plt.title("r={}\nIndices={} and {}".format(mincorrY,ind[1],ind[0]))
plt.scatter(featureY[ind[1]],featureY[ind[0]])
plt.show()
```

However when we scale X by its feature between $[0,1]$, the covariance and correlation changes. Some of the covariance and correlation even has a negative value. We think by scaling the values to just $[0,1]$, we can see more clearly that not all of the variables has a positive association. In that sense, we can say that the covariance and correlation of Y are giving a more accurate data about the relationships between each variable.
